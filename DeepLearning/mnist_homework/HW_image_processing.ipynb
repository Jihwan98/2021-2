```python
import cv2
```

```python
def read_img(path):
    img = cv2.imread(path)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    plt.imshow(img_gray, cmap='gray')
    return img_gray
```

```python
def resize_28(img):
    ret1, thr = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
    ret = cv2.resize(thr, dsize=(28,28), interpolation=cv2.INTER_AREA)
    plt.imshow(ret, cmap='gray')
    return ret
```

```python
img_path = 'images/jh/numbers.png'
numbers_gray = read_img(img_path)
```

```python
numbers_gray.shape
```

```python
img = numbers_gray.copy()
```

```python
num_zero = img[200:250,180:232]
plt.imshow(num_zero, cmap='gray')
```

```python
resize_num_zero = resize_28(num_zero)
```

```python
cv2.imwrite('images/jh/0.jpg', resize_num_zero)
```

```python
img = read_img('images/gy/numbers.jpg')
```

```python
x, y, w, h = cv2.selectROI('img', img, False)
if w and h:
    print(f'x:{x: 3d},y:{y: 3d},w:{w: 3d},h:{h: 3d}')
    roi = img[y:y+h, x:x+w]
    cv2.imshow('roi', roi)
    cv2.moveWindow('roi', 0,0)
    cv2.imwrite('roi.jpg',roi)

cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()

num = read_img('roi.jpg')
resize_num = resize_28(num)
```
